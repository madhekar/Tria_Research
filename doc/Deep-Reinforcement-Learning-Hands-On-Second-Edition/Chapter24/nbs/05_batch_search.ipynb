Check the effect of batched search with constant C and steps

```python
df = pd.concat({
    'b=1':   pd.read_csv("../csvs/t4-c2x2-mcts-c=100-steps=100k.csv"),
    'b=10':  pd.read_csv("../csvs/t4-c2x2-mcts-c=100-steps=100k-b10.csv"),
    'b=100': pd.read_csv("../csvs/t4-c2x2-mcts-c=100-steps=100k-b100.csv"),
})
df.index.set_names(['batch', 'idx'], inplace=True)
df.reset_index(level=0, inplace=True)
df['speed'] = df.solve_steps/df.duration
```

```python
sns.lineplot('depth', 'is_solved', data=df, hue='batch');
```

```python
sns.lineplot('depth', 'solve_steps', data=df, hue='batch');
```

```python
sns.lineplot('depth', 'duration', data=df, hue='batch');
```

```python
sns.lineplot('depth', 'speed', data=df[df.depth > 5], hue='batch');
```

```python
sns.lineplot('depth', 'speed', data=df[df.depth > 35], hue='batch');
```
